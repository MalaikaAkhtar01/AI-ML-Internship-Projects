```python
# ==========================
# TASK 5: AUTO TAGGING SUPPORT TICKETS
# ==========================

!pip install transformers torch pandas

from transformers import pipeline
import pandas as pd

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

labels = [
    "Technical Issue",
    "Billing",
    "Refund",
    "Account Access",
    "Product Inquiry",
    "Delivery Issue"
]

tickets = [
    "I cannot login to my account.",
    "My payment was deducted twice.",
    "I want my money back for cancelled order.",
    "The product has not been delivered yet."
]

results = []

for ticket in tickets:

    output = classifier(
        ticket,
        labels,
        multi_label=True
    )

    top3 = list(zip(
        output["labels"][:3],
        output["scores"][:3]
    ))

    results.append({
        "Ticket": ticket,
        "Top_3_Tags": top3
    })

df = pd.DataFrame(results)

print(df)

# Few-shot example

few_shot_prompt = """
Example 1:
Ticket: Payment deducted twice
Tag: Billing

Example 2:
Ticket: Unable to login
Tag: Account Access

Example 3:
Ticket: Delivery delayed
Tag: Delivery Issue
"""

print("\nFew-Shot Prompt Example:")
print(few_shot_prompt)
```
